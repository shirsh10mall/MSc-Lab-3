```python
import numpy as np
import plotly.express as px
import pandas as pd
import plotly.graph_objects as go
```

```python
# Set 1
readings = pd.DataFrame()
readings['Forward Current I (micro Ampere)'] =  [ 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000, 1500, 2000, 2500, 3000, 4000, 5000, 6000,7000,8000,9000,10000,12000,14000]
readings["ln(I)"] = np.log(readings['Forward Current I (micro Ampere)'])
readings['Junction Voltage (volts)'] = [ 0.603, 0.633, 0.650, 0.662, 0.670, 0.677, 0.683, 0.688, 0.693, 0.696, 0.711, 0.722, 0.730, 0.736, 0.747, 0.755, 0.762,0.767,0.772,0.777,0.780,0.791,0.794]
readings
```

```python
fitted_polynomial_coeff = np.polyfit(readings['Junction Voltage (volts)'],readings['ln(I)'],1)
fitted_polynomial = np.poly1d(fitted_polynomial_coeff)
fitted_data = pd.DataFrame()
fitted_data['x'] = np.linspace(0.6,0.8,num=1000)
fitted_data['y'] = fitted_polynomial(fitted_data['x'])
```

```python
# Create traces
fig = go.Figure()
fig.add_trace(go.Scatter(x=fitted_data['x'], y=fitted_data['y'],mode='lines',name='Fitted Curve'))
fig.add_trace(go.Scatter(x=readings['Junction Voltage (volts)'], y=readings['ln(I)'],mode='markers',name='Observations'))
fig.update_layout(xaxis_title="Junction Voltage (Volts) ------>",yaxis_title="ln(I) ------>",height=600,title = "Exp4.1: Junction Voltage vs ln(I) Graph")
fig.update_layout(legend=dict(
    yanchor="top",
    y=0.99,
    xanchor="left",
    x=0.01
))
```
