```python
import numpy as np
import plotly.express as px
import pandas as pd
import plotly.graph_objects as go
```

```python
readings = pd.DataFrame()
readings['MSR (mm)'] = [0,3,5,7,9]
readings["VSR (mm)"] = [0, 0.65, 0.21, 0.91, 0.14]
readings['Distance (mm)'] = readings['MSR (mm)']+readings['VSR (mm)']
#readings['Distance (mm)'] = readings['Distance (mm)']/1.75
readings['Current (micro-ampere)'] = [ 11.89, 0.0, 0.46, 0.01, 0.1 ]
readings['Order'] = ["Central Maxima", "1st Minima", "1st Maxima", "2nd Minima", "2nd Maxima"]
readings['Position (mm)'] = readings['Distance (mm)'] - readings['Distance (mm)'][0]
readings
```

# Calculations: 

```python
# Theoretical
Lambda = 632.8 * (10**(-9)) # Wavelength
d = 0.29 * (10**(-3)) # Slit Width
# for minima
alpha1_min_Theoretical = np.arcsin(Lambda/d)
alpha2_min_Theoretical = np.arcsin(2*Lambda/d)
# for maxima
alpha1_max_Theoretical = np.arcsin(1.43*Lambda/d)
alpha2_max_Theoretical = np.arcsin(2.459*Lambda/d)
```

```python
# Experimental
a1 = readings["Position (mm)"][1] *(10**(-3)) # Position of 1st Minima
a2 = readings["Position (mm)"][3] *(10**(-3)) # Position of 2nd Minima
a3 = readings["Position (mm)"][2] *(10**(-3)) # Position of 1st Maxima
a4 = readings["Position (mm)"][4] *(10**(-3)) # Position of 2nd Maxima
b = ((125.7+130.7)/2 - 22.2) * (10**(-2)) # Distance b/w Slit and Screen ((120.7+125.7)/2 - 25.7) * (10**(-2))
b
```

```python
# for minima
alpha1_min_Experimental = np.arctan(a1/b)
alpha2_min_Experimental = np.arctan(a2/b)
# for maxima
alpha1_max_Experimental = np.arctan(a3/b)
alpha2_max_Experimental = np.arctan(a4/b)
```
