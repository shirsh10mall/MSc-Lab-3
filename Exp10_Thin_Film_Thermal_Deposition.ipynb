```python
import numpy as np
import plotly.express as px
import pandas as pd
import plotly.graph_objects as go
```

```python
# thickness of film
t = 5.96 * (10^(-6)) # meters

# Print Out Table
readings = pd.DataFrame()

readings['I12 (mA)'] = [2.02, 4.03, 6.06, 8.01, 9.92]
readings['V34 (mV)'] = [1.3, 2.7, 4.0, 5.1, 6.5]

readings["I21 (mA)"] = [2.01, 4.00, 5.98, 8.00, 10.03]
readings['V43 (mV)'] = [1.2, 2.7, 4.0, 5.3, 6.6]

readings['I13 (mA)'] = [2.01, 4.01, 6.0, 8.0, 10.01]
readings['V24 (mV)'] = [1.3, 2.6, 3.8, 5.0, 6.3]

readings['I24 (mA)'] = [2.02, 4.00, 6.02, 8.03, 10.00]
readings['V13 (mV)'] = [1.3, 2.7, 4.1, 5.1, 6.4]

readings
```

```python
fitted_data = pd.DataFrame()

x = readings['V34 (mV)']
y = readings['I12 (mA)']

fitted_data['x Fitted'] = np.linspace(x.min()-0.5,x.max()+0.5, num=2000)

fitted_polynomial_coeff = np.polyfit(x,y,1)
fitted_polynomial = np.poly1d(fitted_polynomial_coeff)
fitted_data['y Fitted'] = fitted_polynomial(fitted_data['x Fitted'])


print("Slope = ", fitted_polynomial_coeff[0] )

# Create traces
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y, mode='markers', name='Observations'))
fig.add_trace(go.Scatter(x=fitted_data['x Fitted'], y=fitted_data['y Fitted'], mode='lines', name='Fitted Curve'))
fig.update_layout(width=1200, height=700 ,legend=dict(yanchor="top", y=0.99, xanchor="left", x=0.01 ))
fig.update_layout( xaxis_title="Voltage V34 (mV)", yaxis_title="Current I12 (mA)", title=" I12 vs V34 Graph" )
fig.update_layout(font=dict( size=18 ))
# fig.update_xaxes(showgrid=True, zeroline=False, gridwidth=0.5, gridcolor='blue', rangemode="tozero")
# fig.update_yaxes(showgrid=True, zeroline=False, gridwidth=0.5, gridcolor='blue', rangemode="tozero")
```
