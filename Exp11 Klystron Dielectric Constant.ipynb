```python
def opt_fun(x, c):
    return (np.tan(x)/x - c)**2
    
# # Create traces
# x = np.linspace(0,10,num=2000)
# fig = px.line(x=x, y=x, title='Plot of Function | y=x')
# fig = px.line(x=x, y=np.tan(x)/x, title='Plot of Function | y = tanx/x')
# fig.show()
```

```python
from scipy.optimize import minimize

const = 1
res = minimize(lambda x: opt_fun(x, const), x0=10)

# Check if the optimization was successful
print(res.success)

# Extract the root from the minimization result
print(res.x)
```
