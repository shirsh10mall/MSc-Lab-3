```python
import numpy as np
import plotly.express as px
import pandas as pd
import plotly.graph_objects as go
pd.set_option('display.max_rows', 500)
```

```python
# Print Out Table
readings = pd.DataFrame()
readings['Current (Ampere)'] = [0, 0.322, 0.514, 1.002, 1.5, 2.014, 2.529, 3.018]
readings['Angle (degree)'] = [142.6, 141.5, 141.2, 140.2, 139.6, 139.0, 138.4, 138.1]
readings['Angle Difference (degree)'] = (np.array([142.6, 141.5, 141.2, 140.2, 139.6, 139.0, 138.4, 138.1]) - 142.6 )*(-1)
readings
```

```python
fitted_data = pd.DataFrame()
fitted_data['Current (Ampere) Fitted'] = np.linspace(readings['Current (Ampere)'].min()-0.1,readings['Current (Ampere)'].max()+0.1,num=2000)

fitted_polynomial_coeff = np.polyfit(readings['Current (Ampere)'],readings['Angle Difference (degree)'],1)
fitted_polynomial = np.poly1d(fitted_polynomial_coeff)
fitted_data['Angle Difference (degree) Fitted'] = fitted_polynomial(fitted_data['Current (Ampere) Fitted'])

# Create traces
fig = go.Figure()
fig.add_trace(go.Scatter(x=readings['Current (Ampere)'], y=readings['Angle Difference (degree)'], mode='markers', name='Observations'))
fig.add_trace(go.Scatter(x=fitted_data['Current (Ampere) Fitted'], y=fitted_data['Angle Difference (degree) Fitted'], mode='lines', name='Fitted Curve'))
fig.update_layout(width=1200, height=700 ,legend=dict(yanchor="top", y=0.99, xanchor="left", x=0.01 ))
fig.update_layout( xaxis_title="Current (Ampere)", yaxis_title="Angle Difference (degree)", title="Current (Ampere) vs Angle Difference (degree) Graph | Set 1" )
# fig.update_xaxes(showgrid=True, zeroline=False, gridwidth=0.5, gridcolor='blue', rangemode="tozero")
# fig.update_yaxes(showgrid=True, zeroline=False, gridwidth=0.5, gridcolor='blue', rangemode="tozero")
```
