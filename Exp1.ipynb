```python
import numpy as np
import plotly.express as px
import pandas as pd
import plotly.graph_objects as go
```

# Nickel

```python
# Print Out Table
readings = pd.DataFrame()
readings['Magnetic Field (Guass)'] = [ 34, 70, 110, 158, 202, 252, 302, 348, 389 ] # Gauss
readings['2*Loop Width (mV)'] = [ 112.5, 243.75, 272.5, 300, 315, 337.5, 355.75, 373.0, 385.5 ] # mV
readings['2*Loop Width (mm)'] = readings['2*Loop Width (mV)']*33.333/1000 #mm
readings['Tip_to_Tip_Width (mV)'] = [ 258.4, 656.2, 719.1, 759.9, 775.2, 797.3, 805.8, 817.7, 822.8 ] # mV
readings['2*Intercept (mV)'] = [ 156.4, 482.8, 520.2, 523.6, 538.9, 537.2, 542, 544, 545  ] # mV
readings
```

```python
readings = pd.DataFrame()
readings['Mag_Field'] = [ 34, 70, 110, 158, 202, 252, 302, 348, 389 ] # Gauss
readings['2*loop_width'] = [ 112.5, 243.75, 272.5, 300, 315, 337.5, 355.75, 373.0, 385.5 ] # mV
readings['2*loop_width'] = readings['2*loop_width']*33.3/1000 #mm
readings['tip_to_tip_width'] = [ 258.4, 656.2, 719.1, 759.9, 775.2, 797.3, 805.8, 817.7, 822.8 ] # mV
readings['2*Intercept'] = [ 156.4, 482.8, 520.2, 523.6, 538.9, 537.2, 542, 544, 545  ] # mV
readings
```

```python
fitted_polynomial_coeff = np.polyfit(readings['Mag_Field'],readings['2*loop_width'],6)
fitted_polynomial = np.poly1d(fitted_polynomial_coeff)
fitted_data = pd.DataFrame()
fitted_data['x'] = np.linspace(25,395,num=2000)
fitted_data['y'] = fitted_polynomial(fitted_data['x'])

# straight fitted line
X = [ 252, 302, 348, 389 ] # Gauss
Y = [ 11.249887, 11.858215, 12.433209	, 12.849871] # mm
fitted_polynomial_coeff = np.polyfit(X,Y,1)
fitted_polynomial = np.poly1d(fitted_polynomial_coeff)
fitted_data_sl = pd.DataFrame()
fitted_data_sl['x'] = np.linspace(0,395,num=2000)
fitted_data_sl['y'] = fitted_polynomial(fitted_data['x'])
#

# Create traces
fig = go.Figure()
fig.add_trace(go.Scatter(x=readings['Mag_Field'], y=readings['2*loop_width'],mode='markers',name='Observations'))
fig.add_trace(go.Scatter(x=fitted_data['x'], y=fitted_data['y'],
                    mode='lines',
                    name='Fitted Curve'))

a1 = fitted_data_sl['x'][0]
a2 = fitted_data_sl['y'][0]
print("a2 = ", a2)
fig.add_hline(y=a2,annotation_text="y=8.590191804281847",annotation_position='top',line_dash='dash')


fig.add_trace(go.Scatter(x=fitted_data_sl['x'], y=fitted_data_sl['y'],
                    mode='lines',
                    name='Fitted Straight Line'))
fig.update_layout(xaxis_title="Magnetic Field (Gauss)",yaxis_title="2L*oop Width (mm)",height=850,title = 'Magnetic Field (Gauss) vs 2*Loop Width  (Nickel)')
```

```python
# Create traces
fig = go.Figure()
fig.add_trace(go.Scatter(x=readings['Mag_Field'], y=readings['2*Intercept'],mode='markers+lines',name='Observations'))
fig.add_hline(y=547,annotation_text="y=547",annotation_position='top')
fig.update_layout(xaxis_title="Magnetic Field (Gauss)",yaxis_title="2*Intercept (mV)",width=1000,height=700,
                  title = ' Magnetic Field (Gauss) vs 2*Intercept (Nickel)')
```

```python
# Create traces
fig = go.Figure()
fig.add_trace(go.Scatter(x=readings['Mag_Field'], y=readings['tip_to_tip_width'],mode='markers+lines',name='Observations'))
fig.add_hline(y=835,annotation_text="y=835",annotation_position='top')
fig.update_layout(xaxis_title="Magnetic Field (Gauss)",yaxis_title='Tip to Tip Width (mV)',width=800,height=700,
                  title = ' Tip to Tip Width vs Magnetic Field Graph (Nickel)')
```
